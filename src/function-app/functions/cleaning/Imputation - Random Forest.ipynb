```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.experimental import enable_iterative_imputer

from sklearn.impute import IterativeImputer
import numpy as np

!pip install pandas numpy scikit-learn fancyimpute
```

```python
df = pd.read_csv('dataset2.csv')
df.head(16)
```

```python
# Check for missing values in each column
missing_values = df.isna().sum()

# Print columns with missing values
print("Columns with missing values:")
print(missing_values[missing_values > 0])
```

```python
# Initialize LabelEncoder
le = LabelEncoder()

# Encode the 'CUSTOMER_EDUCATION_LEVEL' column
df['CUSTOMER_EDUCATION_LEVEL'] = le.fit_transform(df['CUSTOMER_EDUCATION_LEVEL'].astype(str))
```

```python
# Initialize IterativeImputer with RandomForestClassifier
imputer = IterativeImputer(estimator=RandomForestClassifier(n_estimators=100, random_state=0), random_state=0)
```

```python
# Reshape the data to ensure it is in the correct format
df_imputed = df.copy()
df_imputed[['CUSTOMER_EDUCATION_LEVEL']] = imputer.fit_transform(df[['CUSTOMER_EDUCATION_LEVEL']])
```

```python
# Decode the 'CUSTOMER_EDUCATION_LEVEL' column
df_imputed['CUSTOMER_EDUCATION_LEVEL'] = le.inverse_transform(df_imputed['CUSTOMER_EDUCATION_LEVEL'].astype(int))
```

```python
print(df_imputed[['CUSTOMER_EDUCATION_LEVEL']].head())
```

```python
# Save the imputed dataset
df_imputed.to_csv('imputed_dataset.csv', index=False)
```
